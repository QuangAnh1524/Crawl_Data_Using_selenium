3.Chuẩn hóa

```python
import re
def normalize_memory(value):
    if isinstance(value, str):
        # Loại bỏ dấu phẩy và dấu chấm, chỉ giữ lại số và chữ
        value = value.replace(',', '').replace('.', '')
        # Thêm dấu cách giữa số và đơn vị, chuyển sang chữ hoa
        value = re.sub(r'(\d+)([a-zA-Z]+)', r'\1 \2', value.upper())

        if 'GB' not in value and 'MB' not in value:
            value = value + ' GB'
        return value
    return value

data["Dung lượng RAM"] = data["Dung lượng RAM"].apply(normalize_memory)
data["Bộ nhớ trong"] = data["Bộ nhớ trong"].apply(normalize_memory)
data["Kích thước màn hình"] = data["Kích thước màn hình"].str.lower()
```

```python
data.to_csv('cellphones_data.csv', index=False)
```

một số sản phẩm không có giá khuyến mãi, khi crawl về sẽ = 0 => thay giá khuyến mãi chính là giá bán ban đầu

```python
import numpy as np
data['Giá khuyến mãi'] = np.where(data['Giá khuyến mãi'] == 0, data['Giá'], data['Giá khuyến mãi'])
data.to_csv('cellphones_data.csv', index=False)  
```

```python
import re

# chuẩn hóa cho cột Pin
def normalize_battery(value):
    value = str(value)
    value = value.replace(',', '').replace('.', '')
    return re.sub(r'(\d)(mAh)', r'\1 mAh', value)
data["Pin"] = data["Pin"].apply(normalize_battery)
```
